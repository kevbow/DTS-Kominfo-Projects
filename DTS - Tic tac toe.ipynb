```python
from random import randrange

board = [[1, 2, 3], [4, 'X', 6], [7, 8, 9]]

def DisplayBoard():
    print('+-------+-------+-------+')
    print('|       |       |       |')
    print('|   ', board[0][0],'   |   ', board[0][1],'   |   ', board[0][2],'   |', sep = '')
    print('|       |       |       |')
    print('+-------+-------+-------+')
    print('|       |       |       |')
    print('|   ', board[1][0],'   |   ', board[1][1],'   |   ', board[1][2],'   |', sep = '')
    print('|       |       |       |')
    print('+-------+-------+-------+')
    print('|   ', board[2][0],'   |   ', board[2][1],'   |   ', board[2][2],'   |', sep = '')
    print('|       |       |       |')
    print('+-------+-------+-------+')

# code for player move
def EnterMove():
    global count
    vali = False
    while vali == False:
        row, col = int(input('Row : ')), int(input('Column : '))
        if board[row][col] in ('X', 'O'):
            print('Choose another place on the board')
            vali = False
        else:
            board[row][col] = 'O'
            count+=1
            vali=True
            print(count)
            VictoryBoard()
    if winner == False:
        DrawMove()
        
# code for computer move    
def DrawMove():
    global count
    comp = False
    while comp == False:
        rowc = randrange(3)
        colc = randrange(3) 
        if board[rowc][colc] in ('X', 'O'):
            comp = False
        else:
            board[rowc][colc]='X'
            count+=1
            comp=True
            print(count)
            VictoryBoard()
    
def MakeListOfFreeFields():
    if winner != True:
        DisplayBoard()
        for i in range(3):
            for j in range(3):
                if board[i][j] not in ('X', 'O'):
                    print('Board ', i,',', j,' can be filled.', sep='')

# code for checking winning condition                
def VictoryBoard():
    global winner
    if count >= 5:
        if board[0][0]==board[0][1]==board[0][2] == 'X' or board[0][0]==board[0][1]==board[0][2] == 'O':    # top horizontal
            DisplayBoard()
            if board[0][0] == 'X':
                print('Computer win!')
            else:
                print('You win!')
            winner = True
        elif board[1][0]==board[1][1]==board[1][2] == 'X' or board[1][0]==board[1][1]==board[1][2] == 'O':    # mid horizontal
            DisplayBoard()
            if board[1][0] == 'X':
                print('Computer win!')
            else:
                print('You win!')
            winner = True
        elif board[2][0]==board[2][1]==board[2][2] == 'X' or board[2][0]==board[2][1]==board[2][2] == 'O':    # lower horizontal
            DisplayBoard()
            if board[2][0] == 'X':
                print('Computer win!')
            else:
                print('You win!')
            winner = True
        elif board[0][0]==board[1][1]==board[2][2] == 'X' or board[0][0]==board[1][1]==board[2][2] == 'O':    # diagonal
            DisplayBoard()
            if board[0][0] == 'X':
                print('Computer win!')
            else:
                print('You win!')
            winner = True
        elif board[0][0]==board[0][1]==board[0][2] == 'X' or board[0][0]==board[0][1]==board[0][2] == 'O':    # left vertical
            DisplayBoard()
            if board[0][0] == 'X':
                print('Computer win!')
            else:
                print('You win!')
            winner = True
        elif board[0][1]==board[1][1]==board[2][2] == 'X' or board[0][1]==board[1][1]==board[2][2] == 'O':    # mid vertical
            DisplayBoard()
            if board[0][1] == 'X':
                print('Computer win!')
            else:
                print('You win!')
            winner = True
        elif board[0][2]==board[1][2]==board[2][2] == 'X' or board[0][2]==board[1][2]==board[2][2] == 'O':    # right vertical
            DisplayBoard()
            if board[0][2] == 'X':
                print('Computer win!')
            else:
                print('You win!')
            winner = True
        elif board[0][2]==board[1][1]==board[2][0] == 'X' or board[0][2]==board[1][1]==board[2][0] == 'O':    # diagonal
            DisplayBoard()
            if board[0][2] == 'X':
                print('Computer Win!')
            else:
                print('You Win!')
            winner = True
        elif count == 9:
            DisplayBoard()
            print("Game Over! It's a tie.")
            winner = True
            
count = 1        
winner = False
while winner == False:
    EnterMove()
    MakeListOfFreeFields()
```
